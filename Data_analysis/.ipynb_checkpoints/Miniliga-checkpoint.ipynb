```python
#IMPORT
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import Dictionaries


gws = Dictionaries.gws_collect(746814)





```

### Povprečje lige točk na GW

```python
skupni_df = pd.concat(list(gws.values()))
povprecje = skupni_df.groupby('event').mean()
povprecje['points'].plot()
#povprecje
```

### Povprečje lige rank na GW

```python
povprecje['overall_rank'].plot()
```

```python
fig, ax = plt.subplots()
prvih_pet = list(gws.values())[0:4]
for team in prvih_pet:
    ax.plot(team['event'], team['points'], label = team.team_name[0])
ax.set_xlabel('Gameweek')
ax.set_ylabel('Points')
ax.set_title('Points per gameweek \n Top 5 teams')
ax.legend()
```
