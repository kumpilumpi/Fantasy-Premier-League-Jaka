# ANALIZA FPL

Želim analizirati igralce v angleški Premier ligi sezoni 2020/21 in njihov vpliv v Fantasy premier ligi.

```python
#IMPORT

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os


#PODATKI

pot_players = os.path.join(os.pardir, 'FPL-vaastav', 'data', '2020-21', 'cleaned_players.csv') #ne razumem zkj path ni ok -> morm dodt os.pardir
df_players = pd.read_csv(pot_players)

```

## Urejanje podatkov

```python
df_players
```

Prvo izločimo vse neaktivne in premalo aktivne igralce. Tiste, ki so v letošnji sezon igrali manj kot 700 minut in osegli manj kot 40 točk. In dodamo še dodatni polji eff_time, ki pove število doseženih točk na 100 min in eff_price število doseženih točk na milijon. <br>

### Spodnja tabela prikazuje 30 igralcev, z največ doseženimi točkami.

```python
df_players.loc[:, 'eff_time'] = (df_players.total_points / df_players.minutes) * 100
df_players.loc[:, 'eff_price'] = (df_players.total_points / df_players.now_cost) * 10
df_aktivni = df_players.loc[(df_players.minutes > 700) & (df_players.total_points > 40)]
df_aktivni.sort_values('total_points', ascending = False).head(30)
```

### Analiza po pozicijah
Zanima nas v katero vrsto igralcevse nam najbolj splača vožiti sredstva, zato pogledamo povprečne vrednosti igralcev po pozicijah:
* DEF - obrambni igralci 
* FWD - napadalni igralci
* GK - golmani
* MID - vezni igralci

Med aktivnimi igralci je število igralcev po pozicijah razporejeno neenakomerno, zato poglejmo povprečne vrednosti 10 najboljših igralcev po poziciji. Izpustimo stolpce, ki so očitno odvisni od pozicij (največ golov bodo prispevali napadalci,...) ali pa nas ne zanimajo (goals, assists, goals_conceded, selected, minutes, selected_by_percent, yellow_cards, red_cards)

```python
branilci = df_aktivni[df_aktivni.element_type == 'DEF'].sort_values('total_points', ascending = False).head(10)
napadalci = df_aktivni[df_aktivni.element_type == 'FWD'].sort_values('total_points', ascending = False).head(10)
veza = df_aktivni[df_aktivni.element_type == 'MID'].sort_values('total_points', ascending = False).head(10)
golmani = df_aktivni[df_aktivni.element_type == 'GK'].sort_values('total_points', ascending = False).head(10)
top_des = pd.concat([branilci,napadalci,golmani,veza])
povprecje = top_des.groupby('element_type').mean()[['total_points', 'minutes', 'bonus', 'bps', 'ict_index', 'now_cost', 'eff_time', 'eff_price']]
```

```python
povprecje
```

```python
povprecje.plot.bar(y = 'total_points', legend = None)
plt.xlabel(None)
plt.ylabel('Points')
```

Opazimo :
* Največ točk dosegajo v povprečju vezisti sledijo napadalci, obrambni igralci, vratarji.
* V tabeli pri stolpcih bonus (število doseženih bonus točk) in bps (index na podlagi katerega se dodeljuje bonus točke) vidimo, da prevladujejo napadalci in vezni igralci.
* Pri kategoriji eff_time (število točk na časovno enoto) opazimo, da dani čas najbolje izkoristijo vezisti in napadalci, saj so pogosteje zamenjeni tekom tekme in tako preživijo na igrišču manj minut.
* Drugače pa pri eff_price (število doseženih točk na vrednost igralca) največjo vrednost dosegajo vratarji nato, branilci, vezisti, napadalci. 

 ### S čim sovpada vrednost total_points
 Bistvo FPL je nabiranje točk. Zato me zanima ali obstaja kakšna korelacija števila točk s katero drugo kategorijo. 
 1. Korelacija točk in vrednosti igralcev

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])

df_akt_def = df_aktivni.loc[(df_aktivni.element_type == 'DEF')]
ax.scatter(df_akt_def['now_cost'],df_akt_def['total_points'], color = 'b', label = 'DEF')
df_akt_mid = df_aktivni.loc[(df_aktivni.element_type == 'MID')]
ax.scatter(df_akt_mid['now_cost'],df_akt_mid['total_points'], color = 'r',label = 'MID')
df_akt_fwd = df_aktivni.loc[(df_aktivni.element_type == 'FWD')]
ax.scatter(df_akt_fwd['now_cost'],df_akt_fwd['total_points'], color = 'g',label = 'FWD')
df_akt_gk = df_aktivni.loc[(df_aktivni.element_type == 'GK')]
ax.scatter(df_akt_gk['now_cost'],df_akt_gk['total_points'],color = 'y', label = 'GK')


ax.set_xlabel('Points')
ax.set_ylabel('Price')
#ax.set_title(??)
ax.legend()

plt.show()
```

Iz zgornjega prikaza je razvidno, da obstaja zelo šibka korelacija med ceno igralcev in številom doseženih točk, kar ni presenetljivo, saj FPL redno posodoblja cene igralcev na njihovo pordukcijo. Predvsem izstopa dejstvo, da ni nobenega poceni igralca, ki dosega veliko točk (ne vidimo nobene točke v zgornjem levem delu mreže). Kar nakazuje, da administratorji FPL zelo dobro napovedujejo pričakovano število točk igralca v sezoni.  

2. Korelacija točk in ICT-indexa
ICT (influence creativity threat) index je indeks supne produktivnosti igralca na vseh odigranih tekmah. Kombinacija statistično pomembnih podatkov, ki jih beleži FPL, in naj bi FPL managerjem pomagale napovedovati učinkovitost igralca.

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])

df_akt_def = df_aktivni.loc[(df_aktivni.element_type == 'DEF')]
ax.scatter(df_akt_def['ict_index'],df_akt_def['total_points'], color = 'b', label = 'DEF')
df_akt_mid = df_aktivni.loc[(df_aktivni.element_type == 'MID')]
ax.scatter(df_akt_mid['ict_index'],df_akt_mid['total_points'], color = 'r',label = 'MID')
df_akt_fwd = df_aktivni.loc[(df_aktivni.element_type == 'FWD')]
ax.scatter(df_akt_fwd['ict_index'],df_akt_fwd['total_points'], color = 'g',label = 'FWD')
df_akt_gk = df_aktivni.loc[(df_aktivni.element_type == 'GK')]
ax.scatter(df_akt_gk['ict_index'],df_akt_gk['total_points'],color = 'y', label = 'GK')


ax.set_xlabel('ICT index')
ax.set_ylabel('Points')
#ax.set_title(??)
ax.legend()

plt.show()
```

Korelacija med ICT indexom in točkami je veliko boljša. Posebej za veziste in napadalce izgleda, da ICT index zelo uspešno napoveduje njihovo produkcijo točk. Medtem ko je za branilce korelacija manj očitna, za vratarje pa te korelacije skoraj da ni.
