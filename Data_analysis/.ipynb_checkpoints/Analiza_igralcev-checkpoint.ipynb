Poglejmo najboljše defenderje po mojem indeksu:
 -> total_points + ((creaticity + threat) / 100) / minutes
 

```python
dobri_def = df_players[(df_players.total_points > 80) & (df_players.element_type == "DEF")]
dobri_def.sort_values(by="total_points")
```

```python
#dodajanje novega stolpca z mojo statistiko
def_pomembno = dobri_def[["first_name", "second_name", "total_points", "minutes", "creativity", "threat",]]
def_pomembno["moj_indeks"] = (def_pomembno.total_points + ((def_pomembno.creativity + def_pomembno.threat) / 100)) /def_pomembno.minutes * 100
def_pomembno
```
