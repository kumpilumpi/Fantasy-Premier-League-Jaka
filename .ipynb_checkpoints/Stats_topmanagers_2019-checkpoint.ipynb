Rad bi pogledal povprečen gw uporabe wildcarda.

```python
wc_data = podatki.loc[podatki["chip"] == "wildcard"]
wc_data
```

Ločimo tabele na prvi wildcard do 21 kroga in drugi, in pogledamo kdaj povprečno sta bila uporabljena.

```python
prvi_wc = wc_data.loc[wc_data["gw"] <= 21]
drugi_wc = wc_data.loc[wc_data["gw"] > 21]
prvi_wc
#drugi_wc
```

```python
povprecje_1 = (prvi_wc["gw"].sum()/len(prvi_wc))
povprecje_2 = (drugi_wc["gw"].sum()/len(drugi_wc))
povprecje_1.round(), povprecje_2.round()
```
